```python
indicator = set({})
for source in info['Indicator Name']:
    indicator.add(source)
indicator
```

```python
years = []
for column in info.columns:
    if '1' in column or '2' in column:
        years += [column]
years
```

```python
info[years].plot()
```

```python
pivot = pd.pivot_table(info,  columns = "Country Code",values = years)
pivot
```

```python
def first_num(num):
    i=0
    first_num =[]
    for column in pivot.columns:
        if i < num:
            first_num += [column]
        i+=1
    return first_num
```

```python
pivot[first_num(20)].plot(figsize = (10,10))
'done'
```
