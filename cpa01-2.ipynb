```python
years = []
for column in info.columns:
    if '1' in column or '2' in column:
        years += [column]
years
```

```python
pivot = pd.pivot_table(info,index = "Country Name", values = years )
pivot
```

```python
pivot2 = pd.pivot_table(info,  columns = "Country Name",values = years)
pivot2
```

```python
def first_num(num):
    i=0
    first_num =[]
    for column in pivot2.columns:
        if i < num:
            first_num += [column]
        i+=1
    return first_num
```

```python
pivot2[first_num(20)].plot(figsize = (10,10))
#plt.plot(pivot)
'done'
```
